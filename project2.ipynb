```python
def convert_like(data):
    if data['Like'] == 'I hate it!-5':
        return -5
    if data['Like'] == 'I love it!+5':
        return 5
    return data['Like']
df['Like'] = df.apply(convert_like, axis=1)
df
```

```python
def clean_data(df):
    df = pd.get_dummies(data = df, columns=["VisitFrequency"], drop_first = False)
    return df
df=clean_data(df)
df
```

```python
data=pd.DataFrame({col: df[col].astype('category').cat.codes for col in df}, index=df.index)
data
```

```python
X = data.drop(['disgusting'], axis=1) 
y = data['disgusting'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train) 
X_test = sc_x.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
